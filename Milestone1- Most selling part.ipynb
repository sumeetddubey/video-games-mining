```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.patches as mpatches
```

```python
vg_df = pd.read_csv('dataset/games.csv',encoding="ISO-8859-1")
```

```python
genre_by_year = (vg_df.groupby(['Year_of_Release', 'Genre'])[["Global_Sales"]].sum())
platform_by_year = (vg_df.groupby(['Year_of_Release', 'Platform'])[["Global_Sales"]].sum())
```

```python
print("Most selling platforms each year")
platform_by_year[platform_by_year['Global_Sales'] == platform_by_year.groupby(level=[0])['Global_Sales'].transform(max)]
```

```python
print("Most selling genres each year")
genre_by_year[genre_by_year['Global_Sales'] == genre_by_year.groupby(level=[0])['Global_Sales'].transform(max)]
```

```python
fig = plt.figure(figsize=(20, 18))

genre_by_year = (vg_df.groupby(['Year_of_Release', 'Genre']).Global_Sales.sum())
genre_by_year.unstack().plot(kind='area',stacked=True, colormap= 'Spectral', grid=False, figsize=(13, 4))
```

```python
fig = plt.figure(figsize=(20, 18))

platform_sales_percentages_by_year1 = (vg_df.groupby(['Year_of_Release', 'Platform']).Global_Sales.sum())
platform_sales_percentages_by_year1.unstack().plot(kind='area',stacked=True, colormap= 'Spectral', grid=False, figsize=(13, 4))
```
